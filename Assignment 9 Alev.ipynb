```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as con
```

```python
P = 1
h=1
im = 1j
def p(p):
    conds = [p < P/2, (p > -P/2) & (p < P/2), p > P/2]
    funcs = [lambda p: 0, lambda p: 1, 
            lambda p: 0]
    return np.piecewise(p, conds, funcs)


y = np.linspace(-1, 1, 100)
plt.plot(y, p(y),)
plt.show()
```

```python
def X(x):
    return np.sqrt((2*h)/(np.pi*P))*(1/(x))*np.sin(P*x/2*h)
y = np.linspace(-25, 25, 100)
plt.plot(y, X(y),) 
plt.show()
```
