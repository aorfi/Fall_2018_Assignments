```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, solve
import scipy
from scipy import integrate
```

```python
t=0.8
def p(v):
    return 8*t/(3*v-1)-3/(v**2)
```

```python
VP=0.3 # starting geuss for vapor pressure 
x = Symbol("x")
x1,x2,x3 = solve(p(x)-VP,rational=False) # solving for the pionts the horizontal line crosses the isotherm
func=lambda x: p(x) #making a function scipy can read 
area=((x2-x1)*p(x2))*(scipy.integrate.quad(func,x1,x2)[0])-((x3-x2)*p(x2))*(scipy.integrate.quad(func,x2,x3)[0])
"""solving for area region A - region B as defined by the book this is the variable mw then I increase my guess for 
vapor pressure (VP) until the area is close enough to 0 
"""
while np.abs(area) >=0.01:
    VP=VP+0.001
    x = Symbol("x")
    func=8*t/(3*x-1)-3/x**2
    x1,x2,x3 = solve(func-VP,rational=False)
    func=lambda x: 8*t/(3*x-1)-3/x**2
    area=(((x2-x1)*p(x2))-(scipy.integrate.quad(func,x1,x2)[0]))-((scipy.integrate.quad(func,x2,x3)[0])-((x3-x2)*p(x2)))
print("Vapour Pressure:", VP) 
```

```python
v = np.arange(0.4,5,0.01)
plt.title("Van der Waal Isotherm for T/Tc=0.8")
plt.xlabel("V/Vc")
plt.ylabel("P/Pc")
plt.plot(v,p(v))
plt.plot(v,VP*(v/v))
plt.ylim(-0.6, 1.5)
plt.show()
```

```python
v = np.arange(0.5, 40,0.01)
C = 8
def g(v):
    return -t*np.log(3*v-1)+t/(3*v-1)-9/(4*v)+C
plt.title("Gibbs Free Energy")
plt.xlabel("P/Pc")
plt.ylabel("G/(NkTc)")
plt.plot(p(v),g(v))
plt.axvline(x=VP, color="red")
plt.ylim(4, 6)
plt.xlim(-0.4, 0.55)
plt.show()
```
