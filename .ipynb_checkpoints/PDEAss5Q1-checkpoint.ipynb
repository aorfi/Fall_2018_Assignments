```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

Part a

```python
A = np.identity(3)*2+np.diag((-1,-1),1)+np.diag((-1,-1),-1)
Ain = np.linalg.inv(A)
B = np.transpose([[-1,0,1]])
V = np.matmul(Ain,B)
print (V)
```

```python
x = [0,1/4,2/4,3/4,1]
y = [-1,-0.5,0,0.5,1]
plt.title("Solution to 1D Laplace ")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,'r.')
plt.plot(x,y,'')
```

Part b

```python
dag1= [1,1,0,1,1,0,1,1]
dag2= [1,1,1,1,1,1]
A = np.identity(9)*-4+np.diag(dag1,1)+np.diag(dag1,-1)+np.diag(dag2,3)+np.diag(dag2,-3)
Ain = np.linalg.inv(A)
B = np.transpose([[1,1,1,1,1,1,1,1,1]])
V = np.matmul(Ain,B)
print (V)
```

```python
z = np.zeros((5,5))
z[1,1:4]=np.transpose(V[6:9])
z[2,1:4]=np.transpose(V[3:6])
z[3,1:4]=np.transpose(V[0:3])
print(z)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x= y = np.linspace(0, 1,5)
ax.plot_wireframe(x, y, z, linewidth=1)
```
