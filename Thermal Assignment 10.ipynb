```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=0.8
def p(v):
    return 8*t/(3*v-1)-3/(v**2)
```

```python
v = np.linspace(0.4, 5)
plt.title("Van der Waal Isotherm for T/Tc=0.8")
plt.xlabel("Volume (v)")
plt.ylabel("Pressure (p)")
plt.plot(v,p(v))
plt.ylim(-0.6, 1)
plt.show()
```

```python
v = np.linspace(0.6, 2.25)
def g(v):
    return -np.log(3*v)+t/(3*v-1)-9/(4*v)
plt.title("Gibbs Free Energy")
plt.xlabel("pressure")
plt.ylabel("Gibbs Free Energy")
plt.plot(p(v),g(v))
plt.ylim(-3, -2.6)
plt.xlim(0.5, 0.55)
plt.show()
```

```python
a=np.linspace(0.35, 0.45)
for i in range(len(a)):
    if np.any(p(v)==a[i]):
         print(v)
    else:
        i==i+1
```
